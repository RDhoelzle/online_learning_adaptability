```python
codes = {
    'Gender':[], #0=Boy,1=Girl
    'Age':[], #0=1-5, 1=6-10, 2=11-15, 3=16-20, 4=21-25, 5=26-30
    'Education Level':[], #0=School, 1=College, 2=University
    'Institution Type':[], #0=Non Government, 1=Government
    'IT Student':[], #0=No, 1=Yes
    'In Town':[], #0=No, 1=Yes
    'Load-shedding':[], #0=Low, 1=High
    'Financial Condition':[], #0=Poor, 1=Mid, 2=Rich
    'Internet Type':[], #0=Mobile Data, 1=Wifi
    'Network Type':[], #0=2G, 1=3G, 2=4G
    'Class Duration':[], #0=0, 1=1-3, 2=3-6
    'Self Lms':[], #0=No, 1=Yes
    'Device':[], #0=Mobile, 1=Tab, 2=Computer
    'Adaptivity Level':[] #0=Low, 1=Moderate, 2=High
}
```

```python
for i in range(0,df.shape[0]):
    #Gender
    if df['Gender'][i] == 'Boy':
        codes['Gender'].append(0)
    else:
        codes['Gender'].append(1)
    
    #Age
    if df['Age'][i] == '1-5':
        codes['Age'].append(0)
    elif df['Age'][i] == '6-10':
        codes['Age'].append(1)
    elif df['Age'][i] == '11-15':
        codes['Age'].append(2)
    elif df['Age'][i] == '16-20':
        codes['Age'].append(3)
    elif df['Age'][i] == '21-25':
        codes['Age'].append(4)
    else:
        codes['Age'].append(5)
    
    #Education Level
    if df['Education Level'][i] == 'School':
        codes['Education Level'].append(0)
    elif df['Education Level'][i] == 'College':
        codes['Education Level'].append(1)
    else:
        codes['Education Level'].append(2)
        
    #Institution Type
    if df['Institution Type'][i] == 'Non Government':
        codes['Institution Type'].append(0)
    else:
        codes['Institution Type'].append(1)
        
    #IT Student
    if df['IT Student'][i] == 'No':
        codes['IT Student'].append(0)
    else:
        codes['IT Student'].append(1)
        
    #Location In Town
    if df['Location'][i] == 'No':
        codes['In Town'].append(0)
    else:
        codes['In Town'].append(1)
        
    #Load-shedding
    if df['Load-shedding'][i] == 'Low':
        codes['Load-shedding'].append(0)
    else:
        codes['Load-shedding'].append(1)
        
    #Financial Condition
    if df['Financial Condition'][i] == 'Poor':
        codes['Financial Condition'].append(0)
    elif df['Financial Condition'][i] == 'Mid':
        codes['Financial Condition'].append(1)
    else:
        codes['Financial Condition'].append(2)
        
    #Internet Type
    if df['Internet Type'][i] == 'Mobile Data':
        codes['Internet Type'].append(0)
    else:
        codes['Internet Type'].append(1)
        
    #Network Type
    if df['Network Type'][i] == '2G':
        codes['Network Type'].append(0)
    elif df['Network Type'][i] == '3G':
        codes['Network Type'].append(1)
    else:
        codes['Network Type'].append(2)
        
    #Class Duration
    if df['Class Duration'][i] == '0':
        codes['Class Duration'].append(0)
    elif df['Class Duration'][i] == '1-3':
        codes['Class Duration'].append(1)
    else:
        codes['Class Duration'].append(2)
        
    #Self Lms
    if df['Self Lms'][i] == 'No':
        codes['Self Lms'].append(0)
    else:
        codes['Self Lms'].append(1)
        
    #Device
    if df['Device'][i] == 'Mobile':
        codes['Device'].append(0)
    elif df['Device'][i] == 'Tab':
        codes['Device'].append(1)
    else:
        codes['Device'].append(2)
        
    #Adaptivity Level
    if df['Adaptivity Level'][i] == 'Low':
        codes['Adaptivity Level'].append(0)
    elif df['Adaptivity Level'][i] == 'Moderate':
        codes['Adaptivity Level'].append(1)
    else:
        codes['Adaptivity Level'].append(2)
        
codes_df = pd.DataFrame(codes)
codes_df.head()
```

```python
sns.heatmap(codes_df.corr(method='kendall'), cmap='mako')
plt.show()
```
